```python
# Choose Overall Qual, Total Bsmt SF, 1st Flr, Gr Liv Area, Garage Cars, Garage Area, Full Bath, Year Built, Year Remod/Add, Mas Vnr Area
```

```python
# Set the predictor variables and the target
X = ames_df[['Overall Qual', '1st Flr SF', 'Gr Liv Area', 'Full Bath', 'Year Built', 'Year Remod/Add']]
y = ames_df['SalePrice']
```

```python
# Instanize the regression model
lr = LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
X_train_scaled
```

```python
cross_val_score(lr, X_train, y_train, cv=5).mean()
```

```python
cross_val_score(lr, X_test, y_test, cv=5).mean()
```

```python
model = lr.fit(X_test, y_test)
```

```python
model.score(X_test, y_test)
```

```python
prediction = lr.predict(X_test)
```
