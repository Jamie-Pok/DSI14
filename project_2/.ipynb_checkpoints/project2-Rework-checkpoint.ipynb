Next we will turn the ordinal variables into numbers with weightage.

```python
# function to convert ratings to numbers
def ratings_to_ordered_num(df):
    
    list1 = ['exter qual', 'exter cond', 'heating qc', 'kitchen qual']
    list2 = ['bsmt qual', 'bsmt cond', 'fireplace qu', 'garage qual', 'garage cond']
    list3 = ['bsmtfin type 1', 'bsmtfin type 2']
    list4 = ['lot shape', 'utilities', 'land slope', 'bsmt exposure', 'electrical', 'functional', 'garage finish', 'paved drive', 'pool qc', 'fence']
    
    for word1 in list1:
        df[word1] = df[word1].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa':2, 'Po':1})
        print(df[word1].value_counts())
    
    for word2 in list2:
        df[word2] = df[word2].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa':2, 'Po':1, 'NA':0})
        print(df[word2].value_counts())
    
    for word3 in list3:
        df[word3] = df[word3].map({'GLD': 6, 'ALQ': 5, 'BLQ':4, 'Rec':3, 'LwQ':2, 'Unf':1, 'NA':0})
        print(df[word3].value_counts())
        
    df['lot shape'] = df['lot shape'].map({'Reg': 4, 'IR1': 3, 'IR2': 2, 'IR3': 1})
     
    df['utilities'] = df['utilities'].map({'AllPub': 4, 'NoSewr': 3, 'NoSeWa': 2, 'ELO': 1})
 
    df['land slope'] = df['land slope'].map({'Gtl': 3, 'Mod': 2, 'Sev': 1})
        
    df['bsmt exposure'] = df['bsmt exposure'].map({'Gd': 4, 'Av': 3, 'Mn':2, 'No':1, 'NA':0})
        
    df['electrical'] = df['electrical'].map({'SBrkr': 4, 'FuseA': 3, 'FuseF':2, 'FuseP':1, 'Mix':0})
       
    df['functional'] = df['functional'].map({'Typ': 8, 'Min1': 7, 'Min2':6, 'Mod':5, 'Maj1':4, 'Maj2':3, 'Sev':2, 'Sal':1})
       
    df['garage finish'] = df['garage finish'].map({'Fin': 3, 'RFn': 2, 'Unf':1, 'NA': 0})
        
    df['paved drive'] = df['paved drive'].map({'Y': 3, 'P': 2, 'N':1})
        
    df['pool qc'] = df['pool qc'].map({'Ex': 4, 'Gd': 3, 'TA':2, 'Fa':1, 'NA':0})
    
    df['fence'] = df['fence'].map({'GdPrv': 4, 'MnPrv': 3, 'GdWo':2, 'MnWw':1, 'NA':0})
    
    for word4 in list4:
        print(df[word4].value_counts())
    
```

```python
# Call a function to turn ratings into numbers with weightage
ratings_to_ordered_num(train_df)
```

# Modelling

For our models, we will compare between linear regression model, lasso regression model and ridge regression model. We will then pick the best performing model as the model to submit.

```python
# Set the predictor variables and the target
X = feature_df
y = final_train_df['saleprice']
```

```python
# Do a train-test-split on the train data
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=42)
```

```python
# Instanize the regression model
lr = LinearRegression()
```

```python
# Instanize the lasso model
lasso = LassoCV(n_alphas=200)
```

```python
# Instanize the ridge model
ridge = RidgeCV(alphas=np.linspace(.1, 10, 100))
```

```python
# Instanize Standard Scaler and fit & transform the model for X train, transform for X valid
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_valid_scaled = ss.transform(X_valid)
```

```python
# Check X trained scaled
X_train_scaled
```

```python
# fit lasso model with X train, y train to get optimized alpha
lasso.fit(X_train_scaled,y_train)
lasso.alpha_
```

```python
# Cross validate and obtain R2 score for lr
cross_val_score(lr, X_train_scaled, y_train, cv=10).mean()
```

```python
# Cross validate and obtain R2 score for lasso
cross_val_score(lasso, X_train_scaled, y_train, cv=10).mean()
```

```python
# Cross validate and obtain R2 score for ridge
cross_val_score(ridge, X_train_scaled, y_train, cv=10).mean()
```

# Model Selection
The best R2 score seems to be the lasso model. We will use the optimized alpha to see if i can get a better score.

```python
# The best model seem to be lasso. We will try to fit lasso model with optimized alpha
lasso2 = LassoCV(n_alphas=982)
```

```python
# Cross validate and obtain R2 score for lasso with optimized alpha
cross_val_score(lasso2, X_train_scaled, y_train, cv=10).mean()
#there is no improvement
```

There is no improvement, hence we will use back the original lasso model.
